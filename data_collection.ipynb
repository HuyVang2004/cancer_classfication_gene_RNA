## Tải dữ liệu

```python
import pandas as pd
import numpy as np
import requests
import gzip
import shutil
```

```python
url1 = "https://legacy.xenahubs.net/download/TCGA.PANCAN.sampleMap/HiSeqV2.gz"
url2 = "https://pancanatlas.xenahubs.net/download/TCGA_phenotype_denseDataOnlyDownload.tsv.gz"


response1 = requests.get(url1)
with open("GDCdata/HiSeqV2.gz", "wb") as file1:
    file1.write(response1.content)

response2 = requests.get(url2)
with open("GDCdata/TCGA_phenotype_denseDataOnlyDownload.tsv.gz", "wb") as file2:
    file2.write(response2.content)
```

```python
with gzip.open('GDCdata/HiSeqV2.gz', 'rb') as f_in:
    with open('GDCdata/HiSeqV2', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)


with gzip.open('GDCdata/TCGA_phenotype_denseDataOnlyDownload.tsv.gz', 'rb') as f_in:
    with open('GDCdata/TCGA_phenotype_denseDataOnlyDownload.tsv', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
        
```

```python
data = 'GDCdata/HiSeqV2'
labels = 'GDCdata/TCGA_phenotype_denseDataOnlyDownload.tsv'
```

```python
df = pd.read_csv(data, sep='\t').transpose()
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
```

```python
df.head()
```

```python
df.drop('Sample', axis=0, inplace=True)
```

```python
label_df = pd.read_csv(labels, sep = '\t')
```

```python
label_df.head()
```

```python
label_df.set_index('sample', inplace=True)
```

```python
print('Shape of data: ', df.shape)
print('Shape of label: ', label_df.shape)
```

```python
data_full = df.join(label_df[['_primary_disease']])
```
